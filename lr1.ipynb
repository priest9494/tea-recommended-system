### Объявление структуры для хранения мер близости

```python
class Proximity:
    euclid = 0
    tree = 0
    corr = 0
    diffs = 0
    
    def __init__(self, e, t, c, d):
        self.euclid = e
        self.tree = t
        self.corr = c
        self.diffs = d
        
    def getLine(self):
        print("Euclid: %f\nTree: %f\nCorr: %f\nDiffs: %i" %(self.euclid, self.tree, self.corr, self.diffs))
```

## Евклидово расстояние [0;2]

#### Нормирование и формирование массива листьев для реализации перебора

```python
maxPrice = 0
maxWeight = 0
maxLeafSize = 0

minPrice = math.inf
minWeight = math.inf
minLeafSize = math.inf

teaArr = []

for node in PreOrderIter(root):
    if hasattr(node, 'props'):
        teaArr.append(node)
        if node.props.price > maxPrice:
            maxPrice = node.props.price
        if node.props.price < minPrice:
            minPrice = node.props.price
        if node.props.weight > maxWeight:
            maxWeight = node.props.weight
        if node.props.weight < minWeight:
            minWeight = node.props.weight
        if node.props.leafSize > maxLeafSize:
            maxLeafSize = node.props.leafSize
        if node.props.leafSize < minLeafSize:
            minLeafSize = node.props.leafSize
print("%d %d %d %d %d %d" % (maxPrice, maxWeight, maxLeafSize, minPrice, minWeight, minLeafSize))
```

```python
priceNorm = maxPrice - minPrice
def normalizePrice(price):
    return (price - minPrice) / priceNorm
```

```python
weightNorm = maxWeight - minWeight
def normalizeWeight(weight):
    return (weight - minWeight) / weightNorm
```

```python
leafNorm = maxLeafSize - minLeafSize
def normalizeLeafSize(leafSize):
    return (leafSize - maxLeafSize) / leafNorm
```

#### Реализация функции Евклидового расстояния

```python
def euclid(vec1, vec2):
    x1 = normalizePrice(vec1.price)
    x2 = normalizePrice(vec2.price)
    y1 = normalizeWeight(vec1.weight)
    y2 = normalizeWeight(vec2.weight)
    z1 = normalizeLeafSize(vec1.leafSize)
    z2 = normalizeLeafSize(vec2.leafSize)
    
    return math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2 + (z1 - z2) ** 2 + (vec1.arom - vec2.arom) ** 2)
```

```python
def diffs(vec1, vec2):
    dif = 0
    pricePerGramm1 = vec1.price / vec1.weight
    pricePerGramm2 = vec2.price / vec2.weight
    
    # Считаем, что есть разница между чаями с разницей более чем 3 рубля за грамм
    if abs(pricePerGramm1 - pricePerGramm2) > 3:
        dif += 1
        
    # Считаем, что есть разница между чаями с разницей весов более 20 грамм
    if abs(vec1.weight - vec2.weight) > 20:
        dif += 1
        
    if abs(vec1.leafSize - vec2.leafSize) != 0:
        dif += 1
        
    if abs(vec1.arom - vec2.arom) != 0:
        dif += 1
        
    if vec1.ferment != vec2.ferment:
        dif += 1
        
    if vec1.leafForm != vec2.leafForm:
        dif += 1
        
    return dif
```
